```python
import pandas as pd

data = {
    'Name': ['Ali', 'Vali', 'Zohid', 'Muhammad', None, 'Shavkat'],
    'Age': [25, None, 30, 35, 40, None],
    'City': ['Tashkent', None, 'Bukhara', 'Samarkand', 'Fergana', 'Nukus']
}

df = pd.DataFrame(data)
#print(df)
df
```

```python
df.dropna()
```

```python
df
```

```python
agemean=df['Age'].mean()
df['Age'].fillna(agemean,inplace=True)
df
```

```python
df['City'].fillna('Other', inplace=True)
df
```

```python
data3 = {
    'Name': ['Ali', 'Veli', 'Zohid', 'Mukhammad', 'Shavkat'],
    'Age': [25, 30, 30, 35, 40],
    'City': ['Tashkent', 'Bukhara', 'Bukhara', 'Samarkand', 'Fergana']
}

df3 = pd.DataFrame(data3)
df3
```
